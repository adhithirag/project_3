```python
# code generated by chat gpt

import json
import glob
import os

#os.chdir('23 years')
directory = os.getcwd()
print(directory)


# Directory containing your JSON files
json_dir_path = os.getcwd()  # Update this path

# Initialize a dictionary to temporarily hold data by year
data_by_year = {}

# List all files in the directory
for file_name in os.listdir(json_dir_path):
    full_file_path = os.path.join(json_dir_path, file_name)
    if os.path.isfile(full_file_path) and file_name.endswith('.json'):
        with open(full_file_path, 'r') as file:
            json_content = json.load(file)
            
            # Extract year from the "time" field, adjust if necessary
            year = "20" + json_content["time"].split("/")[0]
            
            # Accumulate data by year
            if year not in data_by_year:
                data_by_year[year] = []
            data_by_year[year].append(json_content)

# Write separate files for each year
for year, contents in data_by_year.items():
    output_file_path = os.path.join(json_dir_path, f'{year}_data.json')  # Name files as YEAR_data.json
    with open(output_file_path, 'w') as output_file:
        json.dump(contents, output_file, indent=4)
```
